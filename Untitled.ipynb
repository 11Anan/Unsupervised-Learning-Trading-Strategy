```python
def compute_atr(stock_data):
    atr = pandas_ta.atr(
        high = stock_data['High'], 
        low= stock_data['Low'], 
        close= stock_data['Close'],
        length = 14
    )
    # we also normalize the data
    return atr.sub(atr.mean()).div(atr.std())

df['atr'] = df.groupby(level=1, group_keys=False).apply(compute_atr)
    
```

```python
def compute_macd(close):
    macd = pandas_ta.macd(close = close, length=20).iloc[:,0]
    return macd.sub(macd.mean()).div(macd.std())
df['macd'] = df.groupby(level =1, group_keys= False)['Adj Close'].apply(compute_macd)
```

## Calculate Monthly returns for different time horizons as features

```python
g = data.xs('AAPL', level =1)
g  #Consider only the apple stock for example
```

```python
def calculate_returns(df):
    outlier_cutoff = 0.005
    lags= [1, 2, 3, 6, 9, 12]  #month 
    for lag in lags:
        df[f'return_{lag}m'] = (df['Adj Close'] 
                               .pct_change(lag) #%change
                               .pipe(lambda x: x.clip(lower= x.quantile(outlier_cutoff), 
                                                      upper= x.quantile(1-outlier_cutoff))) #removing outliers
                               .add(1)
                               .pow(1/lag)
                               .sub(1)
                              )
    return df
data = data.groupby(level =1, group_keys = False).apply(calculate_returns).dropna()
```

## Fama-French Factors Download and Calculating Rolling factor betas

- Fama-French data is used to estimate the exposure of assets to common risk factors using linear regression
- The 5 Fama-French factors are market risk, size, value, operating profitability, and investment. The emperically explain asset returns and asses the risk return profile of portfolios
- We use pandas_datareader package.

```python
FFfactor_data = web.DataReader('F-F_Research_Data_5_Factors_2x3', 'famafrench', start= '2010')[0].drop('RF', axis = 1)
#RF is removed as we are nor considering the risk free part
#[0] because we only need the monthly data 
FFfactor_data
```

```python
FFfactor_data.index = FFfactor_data.index.to_timestamp()
FFfactor_data
```

```python
#We had beginning of the month data and the factors are in % so we must divide by 100
FFfactor_data= FFfactor_data.resample('M').last().div(100)
FFfactor_data
```

```python
FFfactor_data = FFfactor_data.join(data['return_1m']).sort_index()
FFfactor_data
```

```python
FFfactor_data.xs('AAPL', level=1).head()
```

```python
#Filter stocks without data for more than 10 months
observations = FFfactor_data.groupby(level =1).size()
valid_stocks = observations[observations>10]
FFfactor_data = FFfactor_data[FFfactor_data.index.get_level_values('Ticker').isin(valid_stocks.index)]
FFfactor_data
```

```python
# Rolling Factor Betas using rolling ols

betas = (FFfactor_data.groupby(level =1, group_keys = False).apply(
    lambda x: RollingOLS(endog= x['return_1m'], 
                         exog= sm.add_constant(x.drop('return_1m', axis=1)),
                         window = min(24, x.shape[0]), #2yrs,
                         min_nobs= len(x.columns)+1).fit(params_only =True).params.drop('const', axis=1)))
                         
```

## K Means clustering for monthl

```python
from sklearn.cluster import KMeans

#data = data.drop('cluster', axis=1)

def get_clusters(df):
    df['cluster'] = KMeans(n_clusters=4,
                           random_state=0,
                           init='random').fit(df).labels_
    return df

data = data.dropna().groupby('Date', group_keys=False).apply(get_clusters)

data
```

```python
data
```

```python
def plot_clusters(data):
    cluster_0 = data[data['cluster']==0]
    cluster_1 = data[data['cluster']==1]
    cluster_2 = data[data['cluster']==2]
    cluster_3 = data[data['cluster']==3]

    plt.scatter(cluster_0.iloc[:,5], cluster_0.iloc[:, 1], color = 'red', label = 'cluster_0')
    plt.scatter(cluster_1.iloc[:,5], cluster_1.iloc[:, 1], color = 'green', label = 'cluster_1')
    plt.scatter(cluster_2.iloc[:,5], cluster_2.iloc[:, 1], color = 'blue', label = 'cluster_2')
    plt.scatter(cluster_3.iloc[:,5], cluster_3.iloc[:, 1], color = 'black', label = 'cluster_3')

    plt.legend()
    plt.show()
    return
    
```

```python
plt.style.use('ggplot')
for i in data.index.get_level_values('Date').unique().tolist():
    g = data.xs(i, level =0)
    plt.title(f'date{i}')
    plot_clusters(g)
```
